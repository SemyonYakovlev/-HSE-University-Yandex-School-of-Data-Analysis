```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
from sklearn.metrics import make_scorer
```

```python
abalone = pd.read_csv('abalone.csv')
abalone
```

```python
abalone['Sex'] = abalone['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
abalone.head()
```

```python
X_train, Y_train = abalone[abalone.columns[:-1]], abalone['Rings']
```

```python
kf = KFold(n_splits=5,shuffle = True, random_state = 1)
kf
```

```python
a = make_scorer(r2_score)
kol_vo = []

for i in range(1,51):
    clf = RandomForestRegressor(n_estimators=i, random_state=1)
    if (sum(cross_val_score(clf, X_train, Y_train, cv=kf, scoring = a)/5)) > 0.52:
            kol_vo.append(i)
```
